# Cleaning 

```python
df["Order Date and Time"]=pd.to_datetime(df["Order Date and Time"])
df.info()
```

```python
df["Delivery Date and Time"]=pd.to_datetime(df["Delivery Date and Time"])
df.info()
```

```python
df.head()
```

```python
def extract(value):
    a=str(value).split()
    return a[0]

df["Discounts and Offers"]=df["Discounts and Offers"].apply(extract)
df.head()
```

```python
def removep(value):
    if "%" in value:
        a = value.replace("%","")
        return float(a)
    else: 
        return float(value)
    
df["Discounts and Offers"]=df["Discounts and Offers"].apply(removep)
df.head()
```

```python
df.loc[(df["Discounts and Offers"]<=15),"Discounts and Offers"]= (df["Discounts and Offers"]/100)*df["Order Value"]
df.head()
```

```python
df["Discounts and Offers"]=df["Discounts and Offers"].fillna(0)
df.head()
```

```python
df["Costs"]=df["Delivery Fee"]+df["Discounts and Offers"]+df["Payment Processing Fee"]
df.head()
```

```python
df["profit"]=df["Commission Fee"]-df["Costs"]
df.head()
```

```python
df["profit"].sum()
```

```python
df['Order Date and Time'] = pd.to_datetime(df['Order Date and Time'])
df.set_index('Order Date and Time', inplace=True)
order_value_over_time = df['Order Value'].resample('D').sum()

plt.figure(figsize=(10, 6))
order_value_over_time.plot(kind='line', marker='o', color='purple')
plt.title('Order Value Over Time')
plt.xlabel('Date')
plt.ylabel('Total Order Value')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Order Value'], df['profit'], color='blue', alpha=0.5)
plt.title('Order Value vs. Profit')
plt.xlabel('Order Value')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```
